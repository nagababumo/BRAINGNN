```python
import deepdish as dd
import os
import numpy as np

def check_file_contents(file_path):
    if not os.path.exists(file_path):
        print(f"File {file_path} does not exist.")
        return
    
    data = dd.io.load(file_path)
    print(f"Contents of {file_path}:")
    for key, value in data.items():
        if isinstance(value, np.ndarray):
            print(f"Key: {key}, Type: {type(value)}, Shape: {value.shape}")
            print(f"Value (first 5 elements):\n{value[:1]}\n")
        else:
            print(f"Key: {key}, Type: {type(value)}, Value: {value}")

# Specify the path to the file you want to check
file_path = '/teamspace/studios/this_studio/data/processed_data/raw/arr_3.h5'
check_file_contents(file_path)
```
